```python
emoticons_str = r"""
  (?:
    [:=;] # Eyes
    [oO\-]? # Nose (optional)
    [D\)\]\(\]/\\OpP] # Mouth
  )"""
regex_str = []
regex_str.append(emoticons_str)
regex_str.append(r'<[^>]+>')# HTML tags
regex_str.append(r'(?:&[\w_]+)')
regex_str.append(r'http[s]?://(?:[a-z]|[0-9]|[$-_@.&amp;+]|[!*\(\),]|(?:%[0-9a-f][0-9a-f]))+') # URLs
regex_str.append(r'(?:(?:\d+,?)+(?:\.?\d+)?)') # numbers
regex_str.append(r"(?:[a-z][a-z'\-\,_]+[a-z])") # words with - and '
regex_str.append(r'(?:[\w_]+)') # other words
regex_str.append(r'(?:\S)') # anything else

tokens_re = re.compile(r'('+'|'.join(regex_str)+')', re.VERBOSE | re.IGNORECASE)
emoticon_re = re.compile(r'^'+emoticons_str+'$', re.VERBOSE | re.IGNORECASE)
```

```python
import nltk 
nltk.download('stopwords')
```

```python
punctuation = list(string.punctuation)
stop = stopwords.words('english') + punctuation + ['…','•']
```

```python
def tokenize(s):
 tokens = tokens_re.findall(s)
 return tokens
def cleanReview(token,regex):
 terms_all = [emoji.demojize(term) for term in token if term.lower() not in stop and not regex.match(term)]
 return terms_all
```

```python
tokens = []
for c in csv:
 tokens.append(tokenize(c[2])) # token akan diambil dari elemen array ke-11 pada setiap baris data review
print(tokens[:5])
```

```python
import operator

exclude_str = []
exclude_str.append(emoticons_str)
exclude_str.append(r'<[^>]+>')# HTML tags
exclude_str.append(r'http[s]?://(?:[a-z]|[0-9]|[$-_@.&amp;+]|[!*\(\),]|(?:%[0-9a-f][0-9a-f]))+') # URLs
exclude_str.append(r'(?:(?:\d+,?)+(?:\.?\d+)?)') # numbers

exclude_re = re.compile(r'('+'|'.join(exclude_str)+')', re.VERBOSE | re.IGNORECASE)
all_word=[]

from collections import Counter
count_all = Counter()

for token in tokens:
    cleanToken = cleanReview(token,exclude_re)
    terms_all = [term.lower() for term in cleanToken]
    count_all.update(terms_all)
    for w in terms_all:
        all_word.append(w)
 
print(count_all.most_common(100))

terms_all
all_word
```

```python
nltk.download('averaged_perceptron_tagger')
pos = nltk.pos_tag([w for w in all_word if w])
pos
```

```python
# j = adjective, r = adverb, n=noun and v = verb
allowed_word_types = ["N","J","V","R"]
all_adj=[w[0] for w in pos if w[1][0] in allowed_word_types]
```

```python
#Menyusun distribusi kata berdasarkan kemunculannya
allwords = nltk.FreqDist(all_adj)
#Mengambil 5000 pertama dari distribusi kata berdasarkan frekwensi kemunculannya
word_features = list(allwords.keys())[:5000]
#Memeriksa word_features
print(word_features[:100])
```

```python
document=[]
for w in csv:
    categ=""
    if w[3] in ['positive']:
        categ="pos"
        document.append((w[2],categ))
    if w[3] in ['negative']:
        categ="neg"
        document.append((w[2],categ))
```

```python
def find_features(document):
    words = word_tokenize(document)
    features = {}
    for w in word_features:
        features[w] = (w in words)
    return features
```

```python
nltk.download('punkt')
```

```python
# Menyusun featureset yang berisi feature word dan ada/tidaknya feature tersebut dalam tweet
featuresets = [(find_features(rev), category) for (rev, category) in document]
```

```python
print(featuresets[:1])
```

```python
# menyusun data training dan testing 
trainsize = round(len(featuresets)*0.7)

random.shuffle(featuresets)
training_set = featuresets[:trainsize]
testing_set = featuresets[trainsize:]

j = 0
for f in featuresets:
    j = j+1 
    
print (j)
#print (featuresets)
print ("training set: ",len(training_set))
print ("testing set: ",len(testing_set))
```

```python
# Menyusun model klasifikasi dengan algoritma Naïve Bayesian dan mengukur akurasinya
classifier = nltk.NaiveBayesClassifier.train(training_set)

print("Classifier accuracy percent:",(nltk.classify.accuracy(classifier, testing_set))*100)
classifier.show_most_informative_features(25)
```

```python
# Menampilkan hasil prediksi dengan model naïve Bayesian
tp = 0
tn = 0
fn = 0
fp = 0

for (rev, category) in document[trainsize:]:
    result=classifier.classify(find_features(rev))
    if category == "pos":
        if category == result:
        #print(rev,"-",category,"-",result)
            tp = tp+1
        else: 
            fn = fn+1 
    if category == "neg":
        #print(rev,"-",category,"-",result)
        if category == result: 
            tn = tn+1
        else: 
            fp = fp+1 
            
print("TP = ",tp, " FN =",fn,"TN = ",tn, " FP =",fp)
```

```python
act = []
pred = []

for i, (feats, label) in enumerate(testing_set): 
    observed = classifier.classify(feats)
    act.append(label)
    pred.append(observed)

print (act)
print (pred)
```

```python
from nltk.metrics import ConfusionMatrix
```

```python
matrix = ConfusionMatrix (act,pred)
print (matrix)
```

```python
# Melakukan pemodelan klasifikasi dengan algoritma Logistic Regression, kemudian ukur tingkat akurasinya;

classifierLog = SklearnClassifier (LogisticRegression(solver='liblinear'), sparse = False).train(training_set)
print("Classifier accuracy percent:",(nltk.classify.accuracy(classifierLog, testing_set))*100)
#classifiersvc.show_most_informative_features(25)
```

```python
from nltk.metrics.scores import (precision, recall)
import collections

refsets = collections.defaultdict(set)
testsets = collections.defaultdict(set)

for i, (feats, label) in enumerate(testing_set):
    refsets[label].add(i)
    observed = classifierLog.classify(feats)
    testsets[observed].add(i)

print('Recall:', recall(refsets['pos'], testsets['pos']))
print('Precision:', precision(refsets['pos'], testsets['pos']))
```
